```python
import requests
import numpy as np
import pandas as pd
import os
import tensorflow as tf
```

```python
def download_from_UCI (UCI_url, name, no_of_features, dest='data'):
    response =  requests.get(UCI_url)
    path = os.path.join(os.getcwd() + "/" + dest + "/" + name + ".csv")
    csv = open(path, "w")
    header = "feature," * no_of_features
    csv.write(header[:-1] + "\n")
    for line in response:
        csv.write(line)
    csv.close()    
    print "done downloading and writing " + name
```

```python
# download datasets
download_from_UCI("https://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication.txt", "banknote_authentication", 5)
download_from_UCI("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", "iris", 4)
download_from_UCI("https://archive.ics.uci.edu/ml/machine-learning-databases/letter-recognition/letter-recognition.data", "letter_recognition", 16)
download_from_UCI("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", "wine", 13)
```

```python
def normalize(data, no_features):
    data_norm = tf.nn.l2_normalize(data.as_matrix(), 1)
    sess = tf.Session()
    with sess.as_default():
        data_norm = data_norm.eval()
#     column = ["feature"] * no_features
    data_norm = pd.DataFrame(data_norm)
    return data_norm
```
